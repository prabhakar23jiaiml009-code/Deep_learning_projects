```python
import random
import pandas as pd

data = []

countries = ["India", "India", "India", "Russia", "China"]
devices = ["mobile", "desktop"]

for i in range(500):
    attempts = random.randint(1,6)
    country = random.choice(countries)
    device = random.choice(devices)
    hour = random.randint(0,23)
    vpn = random.choice([0,1])

    # smarter labeling logic
    label = 1 if (attempts > 3 or country != "India" or vpn == 1 or hour < 6) else 0

    data.append([attempts, country, device, hour, vpn, label])

df = pd.DataFrame(data, columns=[
    "attempts","country","device","hour","vpn","label"
])

df.to_csv("logs.csv", index=False)
print("✅ Data Generated")
```

```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier # type: ignore

df = pd.read_csv("logs.csv")

# encode
df['country'] = df['country'].map({"India":0,"Russia":1,"China":2,"US":3})
df['device'] = df['device'].map({"mobile":0,"desktop":1})

X = df.drop("label", axis=1)
y = df["label"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model = XGBClassifier(n_estimators=100, max_depth=4)
model.fit(X_train,y_train)

print("Score:", model.score(X_test,y_test))

joblib.dump(model,"model.pkl")
```
